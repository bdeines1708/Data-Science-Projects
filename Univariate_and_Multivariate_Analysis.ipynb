```python
import pandas as pd  # imports libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds') #loads dataset
```

```python
print(diamonds.head())  #prints first four rows
```

```python
print(diamonds.info()) #prints info about dataset
```

```python
print(diamonds['carat'].describe())  #Univariate analysis of numerical column 'carat' follows
```

```python
plt.figure(figsize=(10, 6))   #prints histogram of data from column carat
sns.histplot(diamonds['carat'], bins=30, kde=True)
plt.title('Histogram of Carat')
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))  #prints boxplot of data from column 'carat'
sns.boxplot(x=diamonds['carat'])
plt.title('Box Plot of Carat')
plt.show()
```

```python
# Univariate analysis of categorical variable 'cut'
print(diamonds['cut'].value_counts())
```

```python
plt.figure(figsize=(10, 6))  #creates bar blot of data variable 'cut'
sns.countplot(x='cut', data=diamonds)
plt.title('Bar Plot of Cut')
plt.xticks(rotation=45)
```

```python
# Pairplot to visualize relationships between carat, depth, table, price for multivariable analysis
sns.pairplot(diamonds, hue='cut', diag_kind='kde')
plt.title('Pairplot of Diamonds Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))  #scatter plot to visualize relationships between carat, price and cut
sns.scatterplot(x='carat', y='price', hue='cut', data=diamonds, alpha=0.7)
plt.title('Scatter Plot of Carat vs. Price by Cut')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  #box plot to show relationships between cut and price
sns.boxplot(x='cut', y='price', data=diamonds)
plt.title('Box Plot of Price by Cut')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  #violin plot to show relationships between cut and price
sns.violinplot(x='cut', y='price', data=diamonds)
plt.title('Violin Plot of Price by Cut')
plt.show()
```
