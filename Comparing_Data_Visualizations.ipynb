```python
import pandas as pd
import matplotlib.pyplot as plt

#load the dataset
file_path = 'countries.csv'
data = pd.read_csv(file_path)

# Filter the data for the year 1952 and continents Asia and Africa
data_1952 = data[data['year'] == 1952]
asia_1952 = data_1952[data_1952['continent'] == 'Asia']['lifeExpectancy']
africa_1952 = data_1952[data_1952['continent'] == 'Africa']['lifeExpectancy']

# Create subplots to compare the life expectancy histograms for Asia and Africa
fig, ax = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Plot the histogram for Asia
ax[0].hist(asia_1952, bins=15, color='blue', edgecolor='black')
ax[0].set_title('Life Expectancy in Asia (1952)')
ax[0].set_xlabel('Life Expectancy')
ax[0].set_ylabel('Frequency')

# Plot the histogram for Africa
ax[1].hist(africa_1952, bins=15, color='green', edgecolor='black')
ax[1].set_title('Life Expectancy in Africa (1952)')
ax[1].set_xlabel('Life Expectancy')

# Show the plots
plt.tight_layout()
plt.show()


```
