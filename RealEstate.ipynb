```python
import pandas as pd
```

```python
data = pd.read_csv('realestate.csv')
```

```python
data.head()
```

```python
data.houseAge.min()
```

```python
data.houseAge.max()
```

```python
"this gets the mean for normalization"
data.houseAge.mean()
```

```python
"this gets the standard deviation for normalization"
data.houseAge.std()
```

```python
houseAge = data.houseAge
```

```python
"This is the equation for normalization"
norm = (houseAge - houseAge.mean()) / houseAge.std()
norm
```

```python
"this is the equation for the minmax deviation between 1 and zero"
minmax = (houseAge - houseAge.min()) / (houseAge.max())
minmax
```

```python
"this sets a new column for the normalization data"
data['standardized_houseAge'] = norm
```
